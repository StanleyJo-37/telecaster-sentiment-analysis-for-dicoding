```python
from transformers import pipeline
import pandas as pd
import numpy as np
```

```python
pipe = pipeline(
  "zero-shot-classification",
  model="MoritzLaurer/DeBERTa-v3-large-zeroshot-v2.0",
  multi_label=True
)
```

```python
labels = [
  "Dislike",
  "Criticism",
  "Technical Inquiry",
  "Neutral",
  "Love"
]

labels_to_indonesian = {
  "Dislike": "Kritik",
  "Criticism": "Tidak Suka",
  "Technical Inquiry": "Pertanyaan Teknis",
  "Neutral": "Netral",
  "Love": "Suka"
}
```

```python
df = pd.read_csv('test.csv')

df.head()

text_arr = df['text'].to_numpy()

batch_size = 2
split_indices = np.arange(batch_size, len(text_arr), batch_size)

text_batches = np.split(text_arr, split_indices)

text_batches
```

```python
sentiment_batches = []

for batch in text_batches:
  res_batch = pipe(text_batches)
```
