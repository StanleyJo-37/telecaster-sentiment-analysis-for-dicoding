```python
import nltk
from nltk.tokenize import sent_tokenize
import pandas as pd
```

```python
nltk.download('punkt')
```

```python
texts = []
sentences = []
```

```python
def is_valid(text: str):
  text_length = len(text)
  satisfied_length = text_length > 5 && text_length < 40
  
  return satisfied_length

for text in texts:
  tokenized_texts = sent_tokenize(text)
  sentences.extend([t for t in tokenized_texts if is_valid(t)])
```

```python
text_df = pd.DataFrame({
  'texts': sentences,
  'sentiments': ['Undetermined'] * len(sentences)
})
text_df.to_csv('./scraping-results.csv')
```
